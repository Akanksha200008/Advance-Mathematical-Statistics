```python
import pandas as pd
data= pd.read_csv("https://raw.githubusercontent.com/Akanksha200008/Advance-Mathematical-Statistics/refs/heads/main/Police%20Shootings%20Age%20Race.csv")
data.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filtered data for black people
blacks_data = data.query('race == "B"')

# Create a larger figure size
plt.figure(figsize=(10, 6))

# smooth approximation of the histogram with kernel density estimate (KDE)
sns.histplot(data=blacks_data['age'], kde=True, color='RED', edgecolor='black')
plt.xlabel('Age',fontsize=12,fontweight='bold')
plt.ylabel('Frequency',fontsize=12,fontweight='bold')
plt.title('Smooth approximation of age Distribution for Black People with KDE',fontweight='bold')
# Add a grid for better readability
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Use tight layout to avoid cutting off parts of the graph
plt.tight_layout()

plt.show()
```

```python
from scipy.stats import skew, kurtosis

# Filtering data for Black people
blacks_data = data.query('race == "B"')

# Calculating statistics
max_age = blacks_data['age'].max()
min_age = blacks_data['age'].min()

mean_age = blacks_data['age'].mean()
median_age = blacks_data['age'].median()


kurtosis_age = kurtosis(blacks_data['age'])
skewness_age = skew(blacks_data['age'])

# Results
print(f"Maximum Age of Black people: {max_age}")
print(f"Minimum Age of Black people: {min_age}")
print(f"Mean Age of Black people: {mean_age}")
print(f"Median Age of Black people: {median_age}")
print(f"Kurtosis of Age(Blacks): {kurtosis_age}")
print(f"Skewness of Age(Blacks): {skewness_age}")
```

```python
import numpy as np

blacks_data = data.query('race == "B"')
whites_data = data.query('race == "W"')

# Set up the plotting parameters
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")

def cumulative_distribution(data, col):
    sorted_data = np.sort(data[col])
    cumulative_prob = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
    return sorted_data, cumulative_prob

# Calculate cumulative distributions
black_ages, black_cdf = cumulative_distribution(blacks_data, 'age')
white_ages, white_cdf = cumulative_distribution(whites_data, 'age')

# Set up the plotting parameters
plt.figure(figsize=(11, 7))
sns.set(style="whitegrid")

# Creating cumulative distribution plots for Black and White people's ages
plt.plot(black_ages, black_cdf, label='Black', color='purple')
plt.plot(white_ages, white_cdf, label='White', color='green')

# Set plot labels and title
plt.xlabel('Age', fontsize=12, fontweight='bold')
plt.ylabel('Cumulative Probability', fontsize=12, fontweight='bold')
plt.title('Comparison of Cumulative Distributions of Ages between Black and White People', fontsize=12, fontweight='bold')

# Show the plot
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.stats import ttest_ind
# Separate data for Black and White people
blacks_data = data.query('race == "B"')
whites_data = data.query('race == "W"')

# Calculating mean and standard deviation for each group
blacks_std_dev = blacks_data['age'].std()
whites_std_dev = whites_data['age'].std()
blacks_mean_age = blacks_data['age'].mean()
whites_mean_age = whites_data['age'].mean()

# Calculating Cohen's d using formula
pooled_std_deviation = np.sqrt(((len(blacks_data) - 1) * blacks_std_dev**2 + (len(whites_data) - 1) * whites_std_dev**2) / (len(blacks_data) + len(whites_data) - 2))
cohen_d = (whites_mean_age - blacks_mean_age) / pooled_std_deviation

# Display Cohen's d
print(f"Cohen's d value: {cohen_d}")
```

```python
# Performing t-test
t_statistic, p_value = ttest_ind(blacks_data['age'], whites_data['age'], equal_var=False)
print(f"T-Statistic value: {t_statistic}")
print(f"P-Value: {p_value}")
# Checking for statistical significance, common significance level is 0.05
alpha = 0.05

if p_value < alpha:
    print("The mean age difference is statistically significant.")
else:
    print("The mean age difference is not statistically significant.")

```

```python
# Estimation of statistical significance using T-Test Method where the data is observed to be not normally distributed.
black_age_data = data.loc[data['race'] == 'B', 'age']
white_age_data = data.loc[data['race'] == 'W', 'age']

# Calculate the observed test statistic
observed_diff = white_age_data.mean() - black_age_data.mean()

n_permutations = 10000

# Initialize a list to store permutation statistics
permuted_diffs = []

# Permutation test with a modified loop
for _ in range(n_permutations):
    # Shuffle the combined data
    combined_age_data = np.hstack((black_age_data.values, white_age_data.values))
    np.random.shuffle(combined_age_data)

    # Split the permuted data
    permuted_black = combined_age_data[:len(black_age_data)]
    permuted_white = combined_age_data[len(black_age_data):]

    # Append the test statistic difference
    permuted_diffs.append(permuted_white.mean() - permuted_black.mean())

# Convert list to numpy array
permuted_diffs = np.array(permuted_diffs)

# Calculate the p-value
p_val = np.mean(np.abs(permuted_diffs) >= np.abs(observed_diff))

# Print the result
print(f"Observed Statistic: {observed_diff}")
print(f"P-value: {p_val}")

```
